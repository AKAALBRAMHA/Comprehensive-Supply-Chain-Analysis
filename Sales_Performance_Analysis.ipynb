# Sales Performance Analysis
---

### Goal: Aggregate and visualize sales data by product type, SKU, and location.

### Metrics:

  - Total Sales: Number of products sold.
  - Revenue Generated: Total revenue from sales.
  - Production Volumes: Total quantity of products supplied.
  - Customer Demographics: Distribution of sales across different customer demographics.

### Steps:

  - Data Aggregation: Aggregate data by Product Type, SKU, and Location to get total sales and revenue.
  - Visualization:
      - Use bar charts to show total sales and revenue by product type and SKU.
      - Use maps or bar charts to visualize sales performance by location.
      - Pie charts or bar charts can illustrate sales distribution across customer demographics.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/supply_chain_data.csv')
df.head()
```

```python
df.tail()
```

```python
print(f"Number of rows: {df.shape[0]}")
print(f"Number of rows: {df.shape[1]}")
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
sns.heatmap(df.isna().transpose(), cmap = 'Reds')
plt.title('Visualization of Missing Value')
plt.figure(figsize=(10, 8))
plt.show()
```

```python
df.describe()
```

```python
sales_performance = df.groupby(['Product type', 'SKU', 'Location']).agg({
    'Number of products sold': 'sum',
    'Revenue generated': 'sum'
}).reset_index()
```

```python
sns.barplot(data=sales_performance, x='Product type', y='Number of products sold', hue='Location')
plt.title('Total Sales by Product Type and Location')
plt.figure(figsize=(10, 8))
plt.show()
```

```python
sns.barplot(data=sales_performance, x='Product type', y='Revenue generated', hue='Location')
plt.title('Total Revenue by Product Type and Location')
plt.figure(figsize=(10, 8))
plt.show()
```
